```python
#1. How many people have heart disease and how many people dont have heart disease
df.target.value_counts()
```

```python
#plotting a bar chart
df.target.value_counts().plot(kind = 'bar', color=["red", "blue"])
plt.title("Heart Disease Values")
plt.xlabel("1 = Heart Disease, 0 = No Heart Disease")
plt.ylabel("Amount");
```

```python
#plotting pie chart for the same question
df.target.value_counts().plot(kind = 'pie', figsize = (8,6))
plt.legend(["Disease", "No Disease"]);
```

```python
#0 represents 'Female'
#1 represents 'Male'

#Sex column case
#0 represents 'No Disease'
#1 represents 'Disease'

#2. How many 'male' and 'female' are in a dataset?
df.sex.value_counts()
```

```python
#plotting a pie chart
df.sex.value_counts().plot(kind = 'pie', figsize = (8,6))
plt.title('Female And Male Ratio')
plt.legend(['Male', 'Female'])
```

```python
#0 represents 'Female'
#1 represents 'Male'

#Sex column case
#0 represents 'No Disease'
#1 represents 'Disease'

#3. People of which sex has most heart disease?
pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x = 'target', data = df, hue = 'sex')
plt.title("Heart Disease Frequency for sex")
plt.xlabel("0 = No Heart Disease, 1 = Heart Disease");
```

```python
#4. People of which sex has which type of chest pain most?
df.cp.value_counts()
```

```python
#plotting bar chart
df.cp.value_counts().plot(kind = 'bar', color = ['blue','red','yellow','green'])
plt.title("Chest pain type vs count");
```

```python
pd.crosstab(df.sex, df.cp)
```

```python
pd.crosstab(df.sex, df.cp).plot(kind = 'bar', color = ['pink','blue','coral','purple'])
plt.title("Type of chest pain for sex")
plt.xlabel('0 = Female, 1 = Male');
```

```python
#5. People with which chest pain are most pron to have heart disease?
pd.crosstab(df.cp, df.target)
```

```python
sns.countplot(x = 'cp', data = df, hue = 'target');
```

```python
#Most of people who has 'type 0' chest pain has less chance of heart disease and we see the opposite for other types.
#taking a look at our age column.

#Creating a distribution plot with normal distribution curve
sns.displot(x = 'age', data = df, bins = 30, kde = True);
```

```python
# 58 - 59 year old people are most in a dataset.
#Plotting another distribution plot for 'Maximum Heart Rate'
sns.displot(x = 'thalach', data = df, bins = 30, kde = True, color = 'purple');
```

```python
#6. Distribution of Cholesterol Levels
plt.figure(figsize=(10, 6))
sns.histplot(df['chol'], bins=20, kde=True, color='green')
plt.title('Cholesterol Levels Distribution')
plt.xlabel('Cholesterol (mg/dl)')
plt.ylabel('Frequency')
plt.show()



            
            
```
